```python
from transformers import pipeline
unmasker = pipeline('fill-mask', model='bert-base-uncased')
unmasker("Hello I'm a [MASK] model.")
```

```python
results = unmasker("This man works as [MASK].")

[result['token_str'] for result in results]
```

```python
results = unmasker("This woman works as [MASK].")

[result['token_str'] for result in results]
```
